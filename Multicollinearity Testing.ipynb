```python
import numpy
import pandas as pd
from pathlib import Path 
import torch
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
parent_dir = Path("anphy_sleep_data") 
parent_dir = parent_dir.joinpath("patient_records")
raw_dir = parent_dir.joinpath("raw")
clean_dir = parent_dir.joinpath("clean")
body_dir = parent_dir.joinpath("only_body")
data = torch.load(clean_dir / f"EPCTL01_dataset.pt",  weights_only=False)
X = data["X"]
y = data["y"]
```

```python
X.shape
```

```python
X_2d = X.transpose(0, 2, 1).reshape(-1, 14)
df = pd.DataFrame(X_2d)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Feature Correlation Heatmap")
plt.savefig("full_data_correlation.png")
plt.show()
```

```python
data = torch.load(body_dir / f"EPCTL01_body_dataset.pt",  weights_only=False)
X = data["X"]
y = data["y"]
X_2d = X.transpose(0, 2, 1).reshape(-1, 10)
df = pd.DataFrame(X_2d)
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Feature Correlation Heatmap")
plt.savefig("only_body_correlation.png")
plt.show()
```
